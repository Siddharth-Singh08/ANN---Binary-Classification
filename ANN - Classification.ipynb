```python
# Importing the libraries

import numpy as np
import pandas as pd
```

```python
# Importing and reading the dataset
df = pd.read_csv("C:/Users/siddh/Desktop/Datasets/Churn_Modelling.csv")
df
```

```python
# Shape of the dataset is 10000 X 14, where 1 column is the target feature and rest are independent features.
```

```python
# For Predicting the Target feature -(Exited) we will be taking columns from Credit Score to Estimated Salary into consideration
# and split the data accordingly
```

```python
# Splitting the dataset into X(independant) and y(dependant) features.
X = df.iloc[:,3:13]
y = df.iloc[:,13]
```

```python
# Checking the split
X , y
```

```python
# Checking the data types of X
X.dtypes
```

```python
# Features - Geography and Gender are of object type and need to be One-Hot_Encoded(O.H.E) as our ML or DL model will not be 
# able to recognize these categorical/object features
```

```python
# Creating dummy variables
geography = pd.get_dummies(X['Geography'],drop_first = True)
gender = pd.get_dummies(X['Gender'],drop_first = True)
```

```python
# Checking geography and gender O.H.E features
geography, gender
```

```python
# Concatenating the Dataframes
X = pd.concat([X,gender,geography], axis=1)
```

```python
# Checking the new updated dataframe 
X
```

```python
# Dropping Grography and Gender columns without O.H.E 
X = X.drop(['Geography','Gender'], axis=1)
```

```python
# Final Dataframe X for model training
X
```

```python
# Splitting the dataset into Training and Test Set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Checking the train and test split
X_train.shape, X_test.shape, y_train.shape, y_test.shape
# 80% data taken for training and rest 20% for testing
```

```python
# Feature Scaling 

# NOTE: We are doing feature scaling to increase the processing speed of our model.
# 1. As all the features will be scaled, the time taken to compute front and back propagation will reduce.
# 2. The computational time and effort to find the derivative of a scaled number is also less.
# 3. The convergence to the global minima will take lesser time as well because of above reasons mentioned

from sklearn.preprocessing import StandardScaler
s = StandardScaler()
X_train = s.fit_transform(X_train)
X_test = s.transform(X_test)
```

```python
# Making the ANN

# Importing important libraries
import tensorflow as tf
from keras.layers import Dense
from keras.layers import Dropout 
```

```python
# Initializing the ANN
model = tf.keras.Sequential()
```

```python
# Adding the input layer and first hidden layer
model.add(Dense(6, activation ='relu', input_dim = 11))
# Adding the 2nd hidden layer
model.add(Dense(6, activation ='relu'))
# Adding the output layer
model.add(Dense(1, activation = 'sigmoid'))
```

```python
model.summary()
```

```python
# Compiling the ANN
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
# Fitting the ANN to training set 
history = model.fit(X_train,y_train,validation_data=(X_test,y_test),batch_size=100, epochs=200)
```

```python
# Making predictions from the model
y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
# Accuracy Calculation
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_pred,y_test)
acc
```

```python
# Conclusion : Our ANN model has an accuracy of 86.1%. Which means its able to predict correctly whether a person will exit or  
# not exit the bank 86.1 of the times out of the 100% data.
```
